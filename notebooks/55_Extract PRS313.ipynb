## Try to find markers 

```python
import pandas as pd
import subprocess
from pathlib import Path

# -----------------------
# Inputs
# -----------------------
SNP_LIST = "prs313_b38.csv"                 # ton fichier SNP (chr pos ref alt, sans header)
PVAR_TEMPLATE = "acaf_threshold.chr{n}.pvar"  # tes pvar: chr1..chr22
OUT_CSV = "snps_vs_pvar_grep_join.csv"

# -----------------------
# Load SNPs
# -----------------------
snps = pd.read_csv(
    SNP_LIST,
    sep=r"\s+",
    header=None,
    names=["chr", "pos", "ref", "alt"],
    dtype=str
)
snps["chr"] = snps["chr"].astype(str)
snps["pos"] = snps["pos"].astype(str)

# -----------------------
# Build targets file: "chr<TAB>pos<TAB>ref<TAB>alt"
# (awk will load this and print: snp_line ||| pvar_line)
# -----------------------
targets_path = Path("targets_all_chr1_22.tsv")
snps.apply(lambda r: f'{r["chr"]}\t{r["pos"]}\t{r["ref"]}\t{r["alt"]}', axis=1) \
    .to_csv(targets_path, index=False, header=False)

# -----------------------
# Scan all pvars (1 pass per chromosome) and write joined lines
# Output is a "CSV" with 2 columns: snp_side, pvar_side
# (we use a safe delimiter "|||")
# -----------------------
out_path = Path(OUT_CSV)
with out_path.open("w", encoding="utf-8") as out:
    out.write("snp_side,pvar_side\n")  # header

    for n in range(1, 23):
        pvar_file = Path(PVAR_TEMPLATE.format(n=n))
        if not pvar_file.exists():
            print(f"[WARN] Missing: {pvar_file} (skipped)")
            continue

        # awk:
        # - FS split on tabs OR spaces
        # - load targets into dict with key "chr pos"
        # - scan pvar and if key matches print: <snp_line> ||| <pvar_line>
        awk_script = r'''
        BEGIN{FS="[ \t]+"; OFS="\t"}
        NR==FNR{
          k=$1 FS $2;
          snp[k]=$0;
          next
        }
        {
          k=$1 FS $2;
          if (k in snp) {
            # print as two CSV columns (quoted, with internal quotes escaped)
            gsub(/"/, "\"\"", snp[k]);
            line=$0; gsub(/"/, "\"\"", line);
            print "\"" snp[k] "\",\"" line "\""
          }
        }
        '''

        cmd = ["awk", awk_script, str(targets_path), str(pvar_file)]
        res = subprocess.run(cmd, capture_output=True, text=True)

        if res.returncode != 0:
            raise RuntimeError(f"awk failed on {pvar_file}:\n{res.stderr}")

        out.write(res.stdout)

print(f"Done. Wrote: {OUT_CSV}")
```

```python
%%bash

BASE_URI="gs://fc-aou-datasets-controlled/v8/wgs/short_read/snpindel/acaf_threshold/pgen"
for chr in {1..22}; do
    echo "Downloading chr${chr}..."
    gsutil -u $GOOGLE_PROJECT -m -o GSUtil:show_progress_bar=True cp ${BASE_URI}/acaf_threshold.chr${chr}.pgen .
    gsutil -u $GOOGLE_PROJECT -m -o GSUtil:show_progress_bar=True cp ${BASE_URI}/acaf_threshold.chr${chr}.psam .
    plink2 --pfile acaf_threshold.chr${chr} --extract bed1 prs313_b38_all.csv --make-pgen --out prs313_all_chr${chr}
    rm acaf_threshold.chr${chr}.pgen
done
```

# European cohort

## Selection of samples

```python
df_rye = pd.read_csv("aou_admixture_estimates_rye_v8.Q", sep="\t")
```

```python
list_eur = df_rye[df_rye["eur"] >= 0.8]["research_id"].to_list()
```

```python
df_eur = df_final_cohort[df_final_cohort["B"].isin(list_eur)].copy()
```

```python
df_eur["#IID"] = df_eur["B"]
```

```python
keep_ids_eur = df_eur[["#IID","B"]].rename(columns={"B":"IID"})
```

```python
keep_ids_eur.to_csv("keep_ids_eur.txt", sep="\t", index=False)
```

```python
pheno_bc_eur = df_eur[['B', 'has_BC']].rename(columns={"B":"#IID"})
```

## Obtaining allele frequencies

```python
# CASES
!plink2 \
--pfile filtered_eur_chr10 \
--pheno pheno_bc_eur.csv \
--pheno-name has_BC \
--1 \
--keep-if has_BC == 1 \
--chr 10 --from-bp 121580593 --to-bp 121580593 \
--freq \
--out cases_eur
```

```python
# CONTROLS
!plink2 \
--pfile filtered_eur_chr10 \
--pheno pheno_bc_eur.csv \
--pheno-name has_BC \
--1 \
--keep-if has_BC == 0 \
--chr 10 --from-bp 121580593 --to-bp 121580593 \
--freq \
--out control_eur
```
