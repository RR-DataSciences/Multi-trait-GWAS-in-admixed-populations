## Fusion des jeux de données

### Merge `df_ancestry_final` / `df_rye`

```python
# Fusion des jeux de données RYE et Ancestry sur la clé 'research_id' (jointure interne)
df_merged = df_rye_final.merge(df_ancestry_final, on='research_id', how='inner')
```

### Merge `df_with_breast_cancer` / `df_not_breast_cancer`

```python
# Regroupement des individus avec et sans cancer du sein dans un seul DataFrame
df_bc = pd.concat([df_with_breast_cancer_final, df_not_breast_cancer_final], ignore_index=True)
```

### Merge `df_bc` / `df_bam`

```python
# Ajoute une colonne "FID" en première position du DataFrame `df_bc`, en copiant les valeurs de "person_id"
df_bc.insert(0, "FID", df_bc.person_id)
```

```python
# Fusionne les individus présents dans le fichier .fam avec le DataFrame `df_bc` via la colonne "person_id"
df_bc_fam = df_fam[['B']].merge(df_bc, left_on="B", right_on="person_id", how="inner").drop(columns=['FID','person_id'])
```

### Merge `df_bc_fam` /  `df_merged`

```python
# Fusionne les données de génotypage et d’ancestry avec les données phénotypiques, puis supprime la colonne redondante "research_id"
df_merged_all = df_bc_fam.merge(df_merged, left_on='B', right_on='research_id', how='left').drop(columns=['research_id','gender'])
```

```python
df_merged_all
```

## Traitement des données manquantes

```python
# Affiche le nombre de valeurs manquantes pour chaque colonne du DataFrame fusionné
df_merged_all.isna().sum()
```

```python
# Remplace toutes les valeurs manquantes du DataFrame par 0
df_merged_all_not_nan = df_merged_all.fillna(0)
```

## Features engineering

### Ajout d'une colonne age

```python
# Convertit les dates de naissance en format datetime, puis calcule l'âge à partir de l'année en cours
df_merged_all_not_nan['date_of_birth'] = pd.to_datetime(df_merged_all_not_nan['date_of_birth'])
current_year = datetime.now().year
df_merged_all_not_nan['age'] = current_year - df_merged_all_not_nan['date_of_birth'].dt.year

df_merged_all_not_nan.drop(columns="date_of_birth", inplace=True)
df_merged_all_not_nan
```

```python
col = 'age'

# Déplacer la colonne 'age' à la 4ᵉ position (index 3) du DataFrame
col_data = df_merged_all_not_nan.pop(col)
df_merged_all_not_nan.insert(3, col, col_data)
df_merged_all_not_nan
```

## Exportation des données

```python
# Enregistre le DataFrame final dans un fichier TSV local
destination_filename = 'df_final_cohort.tsv'
df_merged_all_not_nan.to_csv(destination_filename, index=False)

# Récupère le nom du bucket de l'environnement d'exécution
my_bucket = os.getenv('WORKSPACE_BUCKET')

# Copie le fichier TSV local vers le dossier "Data" du bucket Google Cloud
args = ["gsutil", "cp", f"./{destination_filename}", f"{my_bucket}/Data/"]
output = subprocess.run(args, capture_output=True)

# Affiche les éventuels messages d’erreur de la commande gsutil
output.stderr
```
