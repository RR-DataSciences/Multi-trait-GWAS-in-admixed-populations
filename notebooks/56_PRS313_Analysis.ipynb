```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

for origin in origin_list:
    df = merge[merge['dominant_origin'] == origin]
    
    # PRS à tester
    prs_list = ["prs_298", "prs_305", "prs_311"]

    y_true = df["has_bc"]

    plt.figure(figsize=(8,6))

    for prs in prs_list:

        y_score = df[prs]

        # Calcul ROC
        fpr, tpr, thresholds = roc_curve(y_true, y_score)

        # Calcul AUC
        auc = roc_auc_score(y_true, y_score)

        # Plot
        plt.plot(fpr, tpr, label=f"{prs} (AUC = {auc:.3f})")

    # Ligne aléatoire
    plt.plot([0,1],[0,1], linestyle="--", color="grey")

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC curves des différents PRS pour la prédiction du cancer du sein \n {origin}", fontsize=13, weight="bold")

    plt.legend()
    plt.grid(alpha=0.3)

    plt.show()
```

# Fonctions AUC cumulative

```python
from sksurv.metrics import cumulative_dynamic_auc

# Function for plotting cumulative dynamic AUC at different follow-up times
def plot_cumulative_dynamic_auc(ax, survival_train, survival_test, risk_score, label, times, color=None):
    # Calculates the cumulative dynamic AUC and its average AUC over all times
    auc, mean_auc = cumulative_dynamic_auc(survival_train, survival_test, risk_score, times)
    
    # Plot of changes in AUC over time
    ax.plot(times, auc, marker='o', color=color, label=label)
    # Average AUC plotted as a dotted line
    ax.plot(times, [mean_auc] * len(times), linestyle='--', color=color)

    # Labels and captions
    ax.set_xlabel("Années depuis l'enrôlement")
    ax.set_ylabel("AUC dynamique cumulative")
    ax.legend()
```
